```python
# DAI011 CAT2 — AI Chatbot using NLTK
# Author: Irwine Kangacha
# Reg No: 250571dai

!pip install nltk
import nltk
import re
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
def get_intent(message):
    message = message.lower()
    tokens = word_tokenize(message)

    if any(word in tokens for word in ['hi', 'hello', 'hey']):
        return "greeting"
    if any(word in tokens for word in ['ai']):
        return "ai_definition"
    elif "course" in tokens or "dai" in tokens:
        return "course_info"
    # Modified to look for 'office' and either 'open' or 'hours'
    elif "office" in tokens and ("open" in tokens or "hours" in tokens):
        return "office_hours"
    elif "contact" in tokens or "email" in tokens:
        return "contact"
    elif any(word in tokens for word in ['bye', 'goodbye']):
        return "goodbye"
    elif any(word in tokens for word in ['thankyou', 'thanks']):
        return "acknowledgement"
    else:
        return "unknown"
```

```python
def chatbot_response(intent):
    responses = {
        "greeting": "Hello! I'm your AI Course Assistant. How can I help you today?",
        # Added a new entry for 'ai_definition'
        "ai_definition": "AI, or artificial intelligence, is the ability of computer systems to perform tasks that normally require human intelligence, such as learning, problem-solving, and decision-making.",
        "course_info": "DAI011 covers Python programming, data analysis, and introductory machine learning.",
        "office_hours": "Office hours are Monday and friday, 2–4 PM.",
        "contact": "You can reach the lecturer at lecturer@example.aiu.edu.",
        "goodbye": "Goodbye! Keep learning and practicing AI concepts!",
        "acknowledgement": "You're welcome!", # Added new entry for acknowledgement
        "unknown": "I'm not sure about that — could you rephrase or ask something else?"
    }
    return responses[intent]
```

```python
print("=== Welcome to the AI Course Assistant Chatbot ===")
while True:
    user_input = input("You: ")
    if user_input.lower() in ['exit', 'quit', 'bye']:
        print("Bot: Goodbye!")
        break
    intent = get_intent(user_input)
    print("Bot:", chatbot_response(intent))
```
